7. Write a program that checks whether a given word is present inside a sentence (string).
Hint: Use the 'in' keyword, like: if word in sentence:


```python
string = "Hello, World!"
check = input("Enter a letter to check: ")
for i in string:
    if i == check:
        found = True
        break
else:
    found = False

if found:
    print(f"The letter '{check}' is present in the string.")
else:
    print(f"The letter '{check}' is not present in the string.")
```

14. Write a program that takes a number and checks whether it is positive or negative. If it is positive, check whether it is even or odd using a nested if.
Hint: First write the outer if-else for positive/negative, then put another if-else inside the 'positive' block to check even/odd.


```python
number = int(input("Enter a number to check : "))
if number>=0:
    print(f"{number} is a positive number.")
    if number%2==0:
        print(f"{number} is an even number.")
    else:
        print(f"{number} is an odd number.")
else:
    print(f"{number} is a negative number.")
```

15. Write a program that takes a student's marks and first checks if the marks are above or below 40 (pass/fail). If passed, use a nested if to decide the grade (A, B, or C).
Hint: Outer if checks marks >= 40 for pass/fail. Inside the pass block, use nested if-elif-else for grade ranges.


```python
mark = int(input("Enter your mark: "))
if mark >=40:
    print("You passed the exam.")
    if mark >=80:
        print("You got an A grade.")
    elif mark >=60:
        print("You got a B grade.")
    elif mark >=50:
        print("You got a C grade.")
else:
    print("You failed the exam.")
    
```

18. Write a function named count_down(n) that uses a loop inside the function to print numbers from n down to 1.
Hint: Use range(n, 0, -1) to count downward.


```python
def count_down(n):
    for i in range(n, 0, -1):
        print(i)
count_down(10)
```

19. Write a function named sum_of_list(numbers) that takes a list and uses a loop inside the function to calculate and return the total sum of the list.
Hint: Initialize total = 0 inside the function, loop through the list adding each item, then return total.


```python
def sum_of_list(numbers):
    total = 0
    for num in numbers:
        total += num
    return total

numbers = [1, 2, 3, 4, 5]
print("Sum of the list:", sum_of_list(numbers))
```

20. Write a function named print_table(number) that uses a loop inside the function to print the multiplication table of that number from 1 to 10.
Hint: Loop from 1 to 10, and inside the loop print number * i for each i.


```python
def print_table(number):
    for i in range(1, 11):
        print(f"{number} x {i} = {number * i}")

print_table(5)
```
