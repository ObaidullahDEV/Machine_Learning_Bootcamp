Hint: find() returns the index of first occurrence. isdigit() returns False if any char is not a digit.


```python
word = 'Banana'
print(word.find('a'))  # Output: 1 - finds the first occurrence of 'a' and returns its index
print(word.count('a'))  # Output: 3 - counts the occurrences of 'a' in the string
print(word.startswith('Ba'))  # Output: True - checks if the string starts with 'Ba'
print(word.endswith('na'))  # Output: True - checks if the string ends with 'na'
print(word.isdigit())  # Output: False - checks if the string consists of digits only
```

Problem 15:  Create three variables: product = 'Laptop', price = 75000.5, discount = 10. Print a formatted line using all three string formatting styles (f-string, .format(), % operator) showing: 'Product: Laptop | Price: 75000.50 | Discount: 10%'.
Hint: Use :.2f for price formatting in all three styles.


```python
product = 'Laptop'
price = 75000.5
discount = 10

# 1. Using f-string
print(f"Product: {product} | Price: {price:.2f} | Discount: {discount}%")

# 2. Using .format()
print("Product: {} | Price: {:.2f} | Discount: {}%".format(product, price, discount))

# 3. Using % operator
print("Product: %s | Price: %.2f | Discount: %d%%" % (product, price, discount))
```

Problem 18:  Take a score (0 - 100) from the user and print a grade using if-elif-else: 90–100 → 'A+', 80–89 → 'A', 60–79 → 'B', 40–59 → 'C', below 40 → 'F'. Also print 'Pass' or 'Fail' based on score >= 40.
Hint: Use 5-branch if-elif-else for grade, then a separate if-else for Pass/Fail


```python
score= int(input("Enter your score between 0 and 100: "))
if score >= 90:
    print("Grade: A+")
elif score >= 80:
    print("Grade: A")
elif score >= 60:
    print("Grade: B")
elif score >= 40:
    print("Grade: C")
else:
    print("Grade: F")

if score>=40:
    print("Pass")
else:
    print("Fail")
```

Problem 19:  Write a login system. Store a correct username and password in two variables. Take username and password input from the user. Using nested if-else: if username matches → check password → print 'Access Granted' or 'Wrong Password'; if username doesn't match → print 'User Not Found'.
Hint: Outer if checks username. Inner if checks password only if username is correct.


```python
Original_Username = "Admin"
Original_Password = "12345"

input_username = input("Enter username: ")

if input_username == Original_Username :
    input_password = input("Enter password: ")
    if input_password == Original_Password:
        print("Login successful!")
    else:
        print("Incorrect password.")

else:
    print("Username not found.")
```
